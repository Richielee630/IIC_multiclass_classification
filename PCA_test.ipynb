```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

dir = 'C:\\Users\\Richie Lee\\Desktop\\College\\1 San Jose State University\\9 2021 fall semester\\CMPE 188\\hw2&hw3\\photos\\train_set'
train_dir = './datasets/train_set/'

targets = ['buildings','forest','glacier','mountain','sea','street']
train_data = []
#load train set
for target in targets:
    train_path = os.path.join(train_dir,target)
    label = targets.index(target)
    for img in os.listdir(train_path):
        train_imgpath = os.path.join(train_path, img) #train_imgpath = './xxx/xxx/xxx.jpg'
        train_img = cv2.imread(train_imgpath)
        try:
            train_img = cv2.resize(train_img,(150,150)) #**very impoertant step
            train_image = np.array(train_img).flatten()
            train_data.append([train_image,label])
        except Exception as e:
            pass

len(train_data)
```

```python
random.shuffle(train_data)
train_data
```

### train data has been shuffled before here 

```python
features = []
labels = []

for feature, label in train_data:
    features.append(feature)
    labels.append(label)
    
labels
```

```python
np_features = np.array(features)
np_lables = np.array(labels)
np_features = np_features.astype(np.int32)
np_features
```

```python
print('total data shape: ',np_features.shape)
print('total labels shape: ', np_lables.shape)
```

## Dementionality reduction n=3100

```python
from sklearn.decomposition import PCA 
pca_95 = PCA(n_components=3100)
np_features_pca = pca_95.fit_transform(np_features)
```

```python
np_features_pca
```

```python
print('total data shape after PCA: ',np_features_pca.shape)
print('total labels shape: ', np_lables.shape)
```

```python
print('the numbers of features have been dropped by', ((len(np_features[1])-len(np_features_pca[1]))/len(np_features[1]))*100,'%')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(), {
    'C': [2,3,4],
    'kernel': ['rbf']
}, cv=3, return_train_score=False)
clf.fit(np_features_pca, np_lables)
clf.cv_results_
```

```python
import pandas as pd
df = pd.DataFrame(clf.cv_results_)
df[['param_C','param_kernel','mean_test_score']]
```
