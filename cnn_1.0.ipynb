```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

dir = 'C:\\Users\\Richie Lee\\Desktop\\College\\1 San Jose State University\\9 2021 fall semester\\CMPE 188\\hw2&hw3\\photos\\train_set'
train_dir = './datasets/train_set/'

targets = ['buildings','forest','glacier','mountain','sea','street']
train_data = []
#load train set
for target in targets:
    train_path = os.path.join(train_dir,target)
    label = targets.index(target)
    for img in os.listdir(train_path):
        train_imgpath = os.path.join(train_path, img) #train_imgpath = './xxx/xxx/xxx.jpg'
        train_img = cv2.imread(train_imgpath)
        try:
            train_img = cv2.resize(train_img,(150,150)) #**very impoertant step
            train_image = np.array(train_img)#.flatten()
            train_data.append([train_image,label])
        except Exception as e:
            pass

len(train_data)
```

```python
random.shuffle(train_data)
train_data
```

### train data has been shuffled before here 

```python
features = []
labels = []

for feature, label in train_data:
    features.append(feature)
    labels.append(label)
    
labels
```

```python
np_features = np.array(features)
np_lables = np.array(labels)
```

```python
np_features = np_features.astype(np.int32)
```

```python
np_features
```

```python
print('total data shape: ',np_features.shape)
print('total labels shape: ', np_lables.shape)
```

```python
import tensorflow.keras.layers as Layers
import tensorflow.keras.activations as Actications
import tensorflow.keras.models as Models
import tensorflow.keras.optimizers as Optimizer
import tensorflow.keras.metrics as Metrics
import tensorflow.keras.utils as Utils
from keras.utils.vis_utils import model_to_dot
```

```python
model = Models.Sequential()

model.add(Layers.Conv2D(200,kernel_size=(3,3),activation='relu',input_shape=(150,150,3)))
model.add(Layers.Conv2D(180,kernel_size=(3,3),activation='relu'))
model.add(Layers.MaxPool2D(5,5))
model.add(Layers.Conv2D(180,kernel_size=(3,3),activation='relu'))
model.add(Layers.Conv2D(140,kernel_size=(3,3),activation='relu'))
model.add(Layers.Conv2D(100,kernel_size=(3,3),activation='relu'))
model.add(Layers.Conv2D(50,kernel_size=(3,3),activation='relu'))
model.add(Layers.MaxPool2D(5,5))
model.add(Layers.Flatten())
model.add(Layers.Dense(180,activation='relu'))
model.add(Layers.Dense(100,activation='relu'))
model.add(Layers.Dense(50,activation='relu'))
model.add(Layers.Dropout(rate=0.5))
model.add(Layers.Dense(6,activation='softmax'))
```

```python
model.compile(optimizer=Optimizer.Adam(lr=0.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
trained = model.fit(np_features,np_lables,epochs=35,validation_split=0.30) 
```

```python
import matplotlib.pyplot as plot

plot.figure(figsize=(20,13))
plot.plot(np.array(trained.history['accuracy']))
plot.plot(np.array(trained.history['val_loss']))
plot.title('Model error')
plot.ylabel('Error')
plot.xlabel('Epoch')
plot.grid(True)
plot.legend(['accuracy', 'val_loss'], loc='upper left',prop = {'size':25})
# plot.savefig('CNNmodelerror.jpg') 
plot.show()
```

### epoch 16 get the best results

```python
model2 = Models.Sequential()

model2.add(Layers.Conv2D(200,kernel_size=(3,3),activation='relu',input_shape=(150,150,3)))
model2.add(Layers.Conv2D(180,kernel_size=(3,3),activation='relu'))
model2.add(Layers.MaxPool2D(5,5))
model2.add(Layers.Conv2D(180,kernel_size=(3,3),activation='relu'))
model2.add(Layers.Conv2D(140,kernel_size=(3,3),activation='relu'))
model2.add(Layers.Conv2D(100,kernel_size=(3,3),activation='relu'))
model2.add(Layers.Conv2D(50,kernel_size=(3,3),activation='relu'))
model2.add(Layers.MaxPool2D(5,5))
model2.add(Layers.Flatten())
model2.add(Layers.Dense(180,activation='relu'))
model2.add(Layers.Dense(100,activation='relu'))
model2.add(Layers.Dense(50,activation='relu'))
model2.add(Layers.Dropout(rate=0.5))
model2.add(Layers.Dense(6,activation='softmax'))
```

```python
model2.compile(optimizer=Optimizer.Adam(lr=0.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model2.summary()
```

```python
trained2 = model2.fit(np_features,np_lables,epochs=16,validation_split=0.30) 
```

```python
plot.figure(figsize=(20,13))
plot.plot(np.array(trained2.history['accuracy']))
plot.plot(np.array(trained2.history['loss']))
plot.plot(np.array(trained2.history['val_accuracy']))
plot.plot(np.array(trained2.history['val_loss']))
plot.title('Model error')
plot.ylabel('Error')
plot.xlabel('Epoch')
plot.grid(True)
plot.legend(['train_acc', 'train_loss','val_acc','val_loss'], loc='upper left',prop = {'size':25})
# plot.savefig('CNNmodelerror.jpg') 
plot.show()
```
