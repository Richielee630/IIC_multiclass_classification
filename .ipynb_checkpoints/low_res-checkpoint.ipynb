```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random
targets = ['buildings','forest','glacier','mountain','sea','street']
train_data = []
train_dir = './low_res/train_set/'
#load train set
for target in targets:
    train_path = os.path.join(train_dir,target)
    label = targets.index(target)
    for img in os.listdir(train_path):
        train_imgpath = os.path.join(train_path, img) #train_imgpath = './xxx/xxx/xxx.jpg'
        train_img = cv2.imread(train_imgpath)
        try:
            train_img = cv2.resize(train_img,(150,150)) #**very impoertant step
            train_image = np.array(train_img).flatten()
            train_data.append([train_image,label])
        except Exception as e:
            pass

len(train_data)
```

```python
random.shuffle(train_data)
train_data
```

```python
features = []
labels = []

for feature, label in train_data:
    features.append(feature)
    labels.append(label)
    
labels
```

```python
np_features = np.array(features)
np_lables = np.array(labels)
np_features = np_features.astype(np.int32)
np_features
```

```python
print('total data shape: ',np_features.shape)
print('total labels shape: ', np_lables.shape)
```

## Dementionality reduction n=2500

```python
from sklearn.decomposition import PCA 
pca_2500 = PCA(n_components=2500)
np_features_pca = pca_2500.fit_transform(np_features)
```

```python
np_features_pca
```

```python
print('total data shape after PCA: ',np_features_pca.shape)
print('total labels shape: ', np_lables.shape)
```

```python
print('the numbers of features have been dropped by', ((len(np_features[1])-len(np_features_pca[1]))/len(np_features[1]))*100,'%')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(), {
    'C': [2,3,4],
    'kernel': ['rbf']
}, cv=3, return_train_score=False)
clf.fit(np_features_pca, np_lables)
clf.cv_results_
```

```python
import pandas as pd
df = pd.DataFrame(clf.cv_results_)
df[['param_C','param_kernel','mean_test_score']]
```

## Dementionality reduction n=500

```python
from sklearn.decomposition import PCA 
pca_500 = PCA(n_components=500)
np_features_pca_500 = pca_500.fit_transform(np_features)
```

```python
np_features_pca_500
```

```python
print('total data shape after PCA: ',np_features_pca_500.shape)
print('total labels shape: ', np_lables.shape)
```

```python
print('the numbers of features have been dropped by', ((len(np_features[1])-len(np_features_pca_500[1]))/len(np_features[1]))*100,'%')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np_features_pca_500, np_lables, test_size=0.2)
```

```python
model = SVC(C=10)
```

```python
y_train.shape
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, model.predict(X_test),target_names=targets))
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
CM = confusion_matrix(y_test, y_predict)
ax = plt.axes()
sn.set(font_scale=1.4)
sn.heatmap(CM, annot=True,annot_kws={"size": 16},  xticklabels=targets, yticklabels=targets, ax = ax, fmt="d",cmap="YlGnBu")
ax.set_title('Confusion matrix')
plt.show()
```

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random
test_data = []
#load test set
for img in os.listdir(test_dir):
    path = os.path.join(test_dir, img)
    test_img = cv2.imread(path)
    
    try:
        test_img = cv2.resize(test_img,(150,150))
        image = np.array(test_img).flatten()
        test_data.append([image,img])
    except Exception as e:
        pass

len(test_data)
```

```python
test_data
```

```python
test_features = []
image_ids = []

for test_feature, image_id in test_data:
    test_features.append(test_feature)
    image_ids.append(image_id)
```

```python
np_test_features = np.array(test_features)
np_test_ids = np.array(image_ids)
np_test_features = np_test_features.astype(np.int32)
```

```python
np_test_features
```

```python
np_test_ids
```

```python
from sklearn.decomposition import PCA
np_test_features_pca = pca_500.transform(np_test_features)
```

```python
pca_500_svm_model_predict = model.predict(np_test_features_pca)
```
