```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

dir = 'C:\\Users\\Richie Lee\\Desktop\\College\\1 San Jose State University\\9 2021 fall semester\\CMPE 188\\hw2&hw3\\photos\\train_set'
train_dir = './photos/train_set/'

targets = ['buildings','forest','glacier','mountain','sea','street']
train_data = []
#load train set
for target in targets:
    train_path = os.path.join(train_dir,target)
    label = targets.index(target)
    for img in os.listdir(train_path):
        train_imgpath = os.path.join(train_path, img) #train_imgpath = './xxx/xxx/xxx.jpg'
        train_img = cv2.imread(train_imgpath)
        try:
            train_img = cv2.resize(train_img,(150,150)) #**very impoertant step
            train_image = np.array(train_img).flatten()
            train_data.append([train_image,label])
        except Exception as e:
            pass

len(train_data)
```

```python
train_data[10000]
```

```python
random.shuffle(train_data)
train_features = []
train_targets = []
for train_feature, train_target in train_data:
    train_features.append(train_feature)
    train_targets.append(train_target)
    
print(train_data[1230])
print(train_features[1230],train_targets[1230])
train_features[1230]
```

```python
train_features[292].shape
```

```python
test_dir = './photos/test_set/'
test_data = []
#load test set
for img in os.listdir(test_dir):
    path = os.path.join(test_dir, img)
    test_img = cv2.imread(path)
    
    try:
        test_img = cv2.resize(test_img,(150,150))
        image = np.array(test_img).flatten()
        test_data.append([image,img])
    except Exception as e:
        pass
```

```python
test_data
```

```python
test_features = []
image_ids = []

for test_feature, image_id in test_data:
    test_features.append(test_feature)
    image_ids.append(image_id)
```

```python
test_features[1][3]
```

```python
print(test_data[1234])
print(test_features[1234],image_ids[1234])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_features, train_targets, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
```

```python
knn_model.score(X_test, y_test)
```

```python
X_test[1].shape
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=3600)
X_pca = pca.fit_transform(train_features)
X_pca
```

```python
X_pca[1].shape
```

```python
X_train_PCA, X_test_PCA, y_train, y_test = train_test_split(X_pca, train_targets, test_size=0.2)
```

```python
X_train_PCA[0].shape
```

```python
from sklearn.svm import SVC
svc_model = SVC(C=3)
svc_model.fit(X_train_PCA, y_train)
```

```python
svc_model_rbf_score = svc_model.score(X_test_PCA, y_test)
```

```python
print('svc model with C =', svc_model.C , ',kernel =', svc_model.kernel)
print('training dataset size =',len(X_train_PCA),'with',len(X_train_PCA[1]),'features after PCA')
print('model score:',svc_model_rbf_score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, svc_model.predict(X_test_PCA),target_names=targets))
```
