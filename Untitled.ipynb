```python
import os 
import numpy as np
import cv2
import matplotlib.pyplot as plt

train_dir = './datasets/train_set'

labels = ['buildings','forest','glacier','mountain','sea','street']

# data = [[image_features],[image_label]]
data = []

for label in labels:
    folder_path = os.path.join(train_dir,label)
    #folder_path = ./datasets/train_set/labels[label]
    image_label = labels.index(label)
    for img in os.listdir(folder_path):
        image_path = os.path.join(folder_path,img)
        #image_path = './datasets/train_set/labels[label]/xxx.jpg'
        image_features = plt.imread(image_path)
        try:
            #data = [[image_features],[image_label]]
            resize_feature = cv2.resize(image_features,(150,150))
            np_train_features = np.array(resize_feature).flatten()
            data.append([np_train_features,image_label])
        except Execption as e:
            pass
```

```python
len(data)
```

```python
train_features = []
train_labels = []

for train_feature,train_label in data:
    train_features.append(train_feature)
    train_labels.append(train_label)
    
```

```python
train_features
```

```python
np_train_features = np.array(train_features)
np_train_labels = np.array(train_labels)
```

```python
np_train_labels
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np_train_features,np_train_labels,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)

# 
# y_pred = model.predcit(X_test)
# score = y_pred/y_test
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
CM = confusion_matrix(y_test, y_pred)
ax = plt.axes()
sn.set(font_scale = 1.5)
sn.heatmap(CM,annot=True,xticklabels=labels, yticklabels=labels, ax = ax, fmt="d")
plt.show()
```
