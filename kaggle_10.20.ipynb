```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

dir = 'C:\\Users\\Richie Lee\\Desktop\\College\\1 San Jose State University\\9 2021 fall semester\\CMPE 188\\hw2&hw3\\photos\\train_set'
train_dir = './datasets/train_set/'

targets = ['buildings','forest','glacier','mountain','sea','street']
train_data = []
#load train set
for target in targets:
    train_path = os.path.join(train_dir,target)
    label = targets.index(target)
    for img in os.listdir(train_path):
        train_imgpath = os.path.join(train_path, img) #train_imgpath = './xxx/xxx/xxx.jpg'
        train_img = cv2.imread(train_imgpath)
        try:
            train_img = cv2.resize(train_img,(150,150)) #**very impoertant step
            train_image = np.array(train_img).flatten()
            train_data.append([train_image,label])
        except Exception as e:
            pass

len(train_data)
```

```python
random.shuffle(train_data)
train_data
```

### train data has been shuffled before here 

```python
features = []
labels = []

for feature, label in train_data:
    features.append(feature)
    labels.append(label)
    
labels
```

```python
np_features = np.array(features)
np_lables = np.array(labels)
```

```python
np_features = np_features.astype(np.int32)
```

```python
np_features
```

```python
print('total data shape: ',np_features.shape)
print('total labels shape: ', np_lables.shape)
```

## Dementionality reduction

```python
from sklearn.decomposition import PCA 
pca_3600 = PCA(n_components=3600)
np_features_pca = pca_3600.fit_transform(np_features)
```

```python
np_features_pca.shape
```

```python
print('total data shape after PCA: ',np_features_pca.shape)
print('total labels shape: ', np_lables.shape)
```

```python
print('the numbers of features have been dropped by', ((len(np_features[1])-len(np_features_pca[1]))/len(np_features[1]))*100,'%')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np_features_pca, np_lables, test_size=0.2510810433812247)
```

## Applying GridsearchCV & Cross-validation for svm

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'), {
    'C': [1,10,20],
    'kernel': ['rbf','linear']
}, cv=5, return_train_score=False)
clf.fit(np_features_pca, np_lables)
clf.cv_results_
```

```python

clf_C = GridSearchCV(svm.SVC(gamma='auto'), {
    'C': [200,500],
    'kernel': ['rbf']
}, cv=2, return_train_score=False)
clf_C.fit(np_features_pca, np_lables)
clf_C.cv_results_
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
clf_C_135 = GridSearchCV(svm.SVC(), {
    'C': [1,3,5],
    'kernel': ['rbf']
}, cv=3, return_train_score=False)
clf_C_135.fit(np_features_pca, np_lables)
clf_C_135.cv_results_
```

```python
test_dir = './datasets/test_set/'
test_data = []
#load test set
for img in os.listdir(test_dir):
    path = os.path.join(test_dir, img)
    test_img = cv2.imread(path)
    
    try:
        test_img = cv2.resize(test_img,(150,150))
        image = np.array(test_img).flatten()
        test_data.append([image,img])
    except Exception as e:
        pass
```
